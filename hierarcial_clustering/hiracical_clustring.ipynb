# Hierarichal Clustering Implementation

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Import IRIS Dataset

iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
# Apply The PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1])
plt.show()
```

```python
# Agglomerative Clusring 
# To Construct a Dendogram

import scipy.cluster.hierarchy as sc 

# Plot the Dendogram
plt.figure(figsize=(20,7))
plt.title("Dendogram")

# Create Dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendogram')
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
plt.show()
```

```python
# Silhoutte Score
from sklearn.metrics import silhouette_score
```

```python
silhouette = []

for k in range(2,10):
    algo = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
    algo.fit(X_scaled)
    score = silhouette_score(X_scaled, algo.labels_)
    silhouette.append(score)
```

```python
## plotting silhouette score
plt.plot(range(2,10),silhouette)
plt.xticks(range(2,10))
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
